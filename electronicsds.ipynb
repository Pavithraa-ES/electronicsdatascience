```python
#Question 2 : Prepare the data frame for further analysis. Clean it, Mold it in any manner you see fit
# Step 1 check for null values, drop it and cross check
df.isnull().sum()
```

```python
df=df.dropna()
```

```python
# Step 3 rectify column name
df.rename(columns={'Order ID': 'OrderID'})
```

```python
# Step 4 Extract date from order date
df['date']=df['Order Date'].str[:8]
```

```python
# Step 5 extract month from order date
df['month']=df['Order Date'].str[:2]
```

```python
# Step 6 extract house_no ,city and pin code from address
df[['HouseNumber', 'City', 'PinCode']] = df['Purchase Address'].str.split(', ', expand=True)
df.head()

```

###### 

```python
# Step 7 Check for month column for bad data entry and remove it
df.month.value_counts()

```

```python
df = df[df["month"].str[:2] != 'Or']
df.head()
```

```python
#step 8 correct the data types
df['month']=df['month'].astype("int32")
df['Order Date']=pd.to_datetime(df['Order Date'])
df['date']=pd.to_datetime(df['date'])
df.info()
```

```python
# Step 9 make Sales Column (hint Quantity_ordered * Price_each)
df['Quantity Ordered']=df['Quantity Ordered'].astype('int32')
df['Price Each']=df['Price Each'].astype('float64')
df.info()
```

```python
df['Sales']= df['Quantity Ordered']*df['Price Each']
df
```

```python
#Question 3 : Sales Value by Each Month
Monthly_sales = df.groupby("month").Sales.sum()
Monthly_sales
```

```python
#Question 4 Which city had the highest sales overall?
city_sales=df.City.value_counts()
city_sales
```

```python
plt.figure(figsize=(16,6))
sns.countplot(x=df['City'])
plt.xlabel("City Name")
plt.ylabel("Number of Sells")
plt.title("City Vs Sales")
```

```python
#Question 5 Find out the distribution of sales over different times of the day (Morning, Afternoon, Evening, Night) and answer the following for each city.
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')  # Ensure "Order Date" is in datetime format
df['Hour'] = df['Order Date'].dt.hour
```

```python
df.groupby("Hour").count()
```

```python
hours = [hour for hour,df in df.groupby('Hour')]
plt.plot(hours,df.groupby('Hour').count())
plt.grid(True)
plt.xticks(hours)
plt.xlabel('Hours in 24 Hr format')
plt.ylabel('Orders')
plt.show()
```

```python
#Question 6 Highest Product Sold by Amount and Numbers
Highest_orders= df.Product.value_counts()
Highest_sales= df.groupby( "Product").Sales.sum().sort_values(ascending=False)

#plotting both the charts
fig, axes = plt.subplots(1, 2, figsize=(20,7))

sns.barplot(ax=axes[0], x=Highest_orders.index, y=Highest_orders.values)
axes[0].set_xlabel('Products') 
axes[0].set_ylabel('No. of Orders')
axes[0].tick_params(axis='x', rotation=90)

sns.barplot(ax=axes[1],x = Highest_sales.index, y=Highest_sales.values)
axes[1].set_xlabel( 'Products')
axes[1].set_ylabel('Sales Amount')
axes[1].tick_params(axis='x', rotation=90) 
plt.show()
```
